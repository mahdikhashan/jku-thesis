```python
import json
from lm_eval import simple_evaluate
```

```python
results = simple_evaluate(
    model="hf",
    model_args="pretrained=meta-llama/Llama-3.2-1B,dtype=bfloat16",
    tasks=["piqa", "arc_easy", "arc_challenge", "hellaswag", "winogrande"],
    num_fewshot=0,
    batch_size=8,
)
```

```python
import os, pathlib
out_dir = pathlib.Path("./eval_results/llama32_1b_0shot")
out_dir.mkdir(parents=True, exist_ok=True)
with open(out_dir / "results.json", "w") as f:
    # results contains non-serializable bits; drop them
    json.dump({"results": results["results"], "configs": results["configs"]}, f, indent=2, default=str)

for task, metrics in results["results"].items():
    print(task, metrics)
```
